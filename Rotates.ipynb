This version uses magic widget from matplotlib and rotates the graphs for better understanding.

Note that for z = x + y,
−x−y+z=0 is the plane equation, and thus the normal vector of the plane is (-1, -1, 1)
Third quadrant, positive z. That's why the plane is tilted

```python
# Comment the following back in to install magic widget or %matplotlib widget won't run:
#%pip install ipympl
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # needed for 3D plotting

# Create grid of x and y values
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = X + Y  #equation to plot

# Make the figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='plasma', edgecolor='none')

# Add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

#Draw the positive axes
ax.quiver(0, 0, 0, 5, 0, 0, color="red", arrow_length_ratio=.1)  # x-axis #code in form of vector starting at 0,0,0 with red arrow in direction of x axis
ax.quiver(0, 0, 0, 0, 5, 0, color="green", arrow_length_ratio=.1)  # y-axis
ax.quiver(0, 0, 0, 0, 0, 5, color="blue", arrow_length_ratio=.1)  # z-axis

# Draw the negative directional axes
ax.quiver(0, 0, 0, -5, 0, 0, color="red", arrow_length_ratio=.1) #Note that the length 5 is scaled up from 1 to lengthen the axes for visibility
ax.quiver(0, 0, 0, 0, -5, 0, color="green", arrow_length_ratio=.1)
ax.quiver(0, 0, 0, 0, 0, -5, color="blue", arrow_length_ratio=.1)



plt.show()
```

```python

# Grid for x and y
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Different planes (same slope, shifted by z_offset)
z_offsets = [0, 1, 2, ]
colors = ['red', 'green', 'blue']

for z_off, color in zip(z_offsets, colors):
    Z = X**2 + Y**2 + z_off  # shift entire plane
    ax.plot_surface(X, Y, Z, color=color, alpha=0.6)

# Axes arrows
ax.quiver(-6, 0, 0, 12, 0, 0, color="red", arrow_length_ratio=0.05)
ax.quiver(0, -6, 0, 0, 12, 0, color="green", arrow_length_ratio=0.05)
ax.quiver(0, 0, -6, 0, 0, 12, color="blue", arrow_length_ratio=0.05)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([-5,5])
```

Code for visualizing f(x,y) = |y|

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid of x and y values
x = np.linspace(-5, 5, 100)  # x runs from -5 to 5
y = np.linspace(-5, 5, 100)  # y runs from -5 to 5
X, Y = np.meshgrid(x, y)

# Define z = |y|
Z = np.abs(Y)

# Plot
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.9)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Surface z = |y|')

plt.show()
```
